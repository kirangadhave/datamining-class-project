```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LassoCV, LassoLarsCV, LassoLarsIC
from sklearn.cluster import KMeans
import os
```

## Import Player Attributes

```python
attribute_path = os.getcwd() + '/Data/Player_Attributes.csv'
attributes = pd.read_csv(attribute_path).dropna()
first_import = np.array([True if x == '2007-02-22 00:00:00' else False for x in np.array(attributes['date'].values)])
attributes = attributes[first_import]
print(attributes.columns)
f_path = os.getcwd() + '/Data/Player.csv'
players = pd.read_csv(f_path).dropna()
print(players.columns)
players.set_index('player_fifa_api_id', inplace=True)
```

```python
# separate overall attributes and all the numerical player attributes apart
all_y = attributes['overall_rating'].values
all_x = attributes.loc[:,'crossing':'gk_reflexes'].values
attribute_names = attributes.loc[:,'crossing':'gk_reflexes'].columns.values
```

```python
attacking_attributes = ['crossing','finishing','heading_accuracy','short_passing','volleys']
skill_attributes = ['dribbling','curve','free_kick_accuracy','long_passing','ball_control']
movement_attributes = ['acceleration','sprint_speed','agility','reactions','balance']
power_attributes = ['shot_power','jumping', 'stamina','strength','long_shots']
mentality_attributes = ['aggression','interceptions','positioning','vision','penalties']
defending_attributes = ['marking','standing_tackle','sliding_tackle']
goalkeeping_attributes = ['gk_diving','gk_handling','gk_kicking','gk_positioning','gk_reflexes']
attribute_lists = [attacking_attributes,skill_attributes,movement_attributes,power_attributes,\
                   mentality_attributes,defending_attributes,goalkeeping_attributes]
```

```python
def rbv_by_attr(ax, rbv, attribute_names, seperate_attributes, **kwargs):
    """ 
    ax: subplot object to plot on
    rbv: V^T
    vector_num: vector num to plot
    attribute_names: list of attribute names
    separate_attributes: list of attribute names split by attacking, defense etc role
    
    **kwargs: any valid kwarg for plt.plot
    """
     # Make a pandas dataframe
    index = np.arange(len(rbv))
    vals = np.sort(rbv)
    plot_info = np.vstack((vals,index))
    plot_df = pd.DataFrame(plot_info, columns=attribute_names[np.argsort(rbv)])
    for i in range(len(seperate_attributes)):
        ax.plot(plot_df[seperate_attributes[i]].values[1,:],
                plot_df[seperate_attributes[i]].values[0,:],'o',**kwargs)
    
    return
```

## SVD Investigation

```python
U, W, V = np.linalg.svd(all_x, full_matrices=False)
fig, ax = plt.subplots(5,2,figsize=(12,6))
rbv_by_attr(ax[0,0],V[0,:],attribute_names,attribute_lists)
ax[0,1].plot(np.sort(U[:,0]),'o')
rbv_by_attr(ax[1,0],V[1,:],attribute_names,attribute_lists)
ax[1,1].plot(np.sort(U[:,1]),'o')
rbv_by_attr(ax[2,0],V[2,:],attribute_names,attribute_lists)
ax[2,1].plot(np.sort(U[:,2]),'o')
rbv_by_attr(ax[3,0],V[3,:],attribute_names,attribute_lists)
ax[3,1].plot(np.sort(U[:,3]),'o')
rbv_by_attr(ax[4,0],V[4,:],attribute_names,attribute_lists)
ax[4,1].plot(np.sort(U[:,4]),'o')
print('green: movement\nred: power\norange: skill\nblue: attacking\npurple: mentality\nbrown: defending\npink: GK')
```
